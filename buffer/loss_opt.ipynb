```python
import matplotlib.pyplot as plt
def warm_and_decay_lr_scheduler(scheduler_gamma, decay_steps_pct, total_steps, warmup_steps_pct, step):
    warmup_steps = warmup_steps_pct * total_steps
    decay_steps = decay_steps_pct * total_steps
    assert step < total_steps
    if step < warmup_steps:
        factor = step / warmup_steps
    else:
        factor = 1
    factor *= scheduler_gamma ** (step / decay_steps)
    return factor
    
```

```python

lr = 0.0004
warmup_steps_pct = 0.02 # 0.02 # 0.05
decay_steps_pct = 0.2 # 0.2 # 0.3
weight_decay=0.0
max_epochs=100
scheduler_gamma=0.5 # 0.5 # 0.05
total_steps=100

lr_lst = []
for step in range(total_steps):

    lr_lst.append(lr*warm_and_decay_lr_scheduler(step=step, scheduler_gamma=scheduler_gamma,
                    decay_steps_pct=decay_steps_pct,
                    total_steps=total_steps,
                    warmup_steps_pct=warmup_steps_pct))


plt.plot(lr_lst)
```
